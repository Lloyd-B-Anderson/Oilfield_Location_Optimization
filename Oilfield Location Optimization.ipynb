```python
# Importacion de datos 
geo_0 = pd.read_csv('geo_data_0.csv')
geo_1 = pd.read_csv('geo_data_1.csv')
geo_2 = pd.read_csv('geo_data_2.csv')
```

# 2. Entrenamiento de modelo 

```python
# Funcion para predecir con el modelo de regresion lineal 

def pred(df): 
    # Se elimina la columna id ya que no brinda informacion relevante
    # Caracteristicas y objetivo 
    features = df.drop(['product','id'], axis=1) 
    target = df['product']

    # Division de datos en conjuntos de entrenamiento y validacion 
    features_train, features_valid, target_train, target_valid = train_test_split(features,target, test_size = 0.25, random_state = 123)

    # Entrenar modelo de regresion lineal 
    model = LinearRegression() 
    model.fit(features_train, target_train)

    # Predicciones con el conjunto de validacion 
    predict = model.predict(features_valid)

    return predict, target_valid
        
```

```python
# calculo de media de la prediccion y RMSE para geo_0
predict_0, target_valid_0 = pred(geo_0)

avg_volume_0 = predict_0.mean() 
RMSE_0 = root_mean_squared_error(target_valid_0, predict_0)

# calculo de media de la prediccion y RMSE para geo_1
predict_1, target_valid_1 = pred(geo_1)

avg_volume_1 = predict_1.mean() 
RMSE_1 = root_mean_squared_error(target_valid_1, predict_1)

# calculo de media de la prediccion y RMSE para geo_2
predict_2, target_valid_2 = pred(geo_2)

avg_volume_2 = predict_2.mean() 
RMSE_2 = root_mean_squared_error(target_valid_2, predict_2)


print(f'Volumen medio de reservas predicho:\n\n\t- geo_data_0: {avg_volume_0}\n\t- geo_data_1: {avg_volume_1}\n\t- geo_data_2: {avg_volume_2}\n')
print(f'RMSE del modelo para:\n\n\t- geo_data_0: {RMSE_0}\n\t- geo_data_1: {RMSE_1}\n\t- geo_data_2: {RMSE_2}')
```

# 3. Calculo de ganancia

```python
budget = 100e6 # Presupuesto total
pozos = 200 # Pozos por seleccionar
ingreso_barril = 4500 # Ingreso por mil barriles 
ingreso_necesario_pozo = 500e3 # Ingreso requerido por hoyo 
risk_threshold = 2.5 # Umbral maximo de riesgo permitido 
n_top = 200 #primeros 200 mejores resultados 
```

```python
# Funcion para seleccionar los puntos de la region con mayor volumen de reservas segun el modelo 
def selec_hoyos(predictions):
    sorted_index = np.argsort(predictions)[::-1] 
    top_index = sorted_index[:n_top]
    top_predictions = predictions[top_index]
    
    return top_index, top_predictions

# Funcion para calcular el beneficio
def calc_profit(total_predicted_product):
    revenue = total_predicted_product * ingreso_barril
    profit = revenue - budget 
    return profit 
```

```python
# Conseguir la reserva de medias predichas para los 200 pozos para geo_0
top_index_0, top_predictions_0 = selec_hoyos(predict_0)
avg_top_pred_0 = top_predictions_0.mean()

# Conseguir la reserva de medias predichas para los 200 pozos para geo_1
top_index_1, top_predictions_1 = selec_hoyos(predict_1)
avg_top_pred_1 = top_predictions_1.mean()

# Conseguir la reserva de medias predichas para los 200 pozos para geo_2
top_index_2, top_predictions_2 = selec_hoyos(predict_2)
avg_top_pred_2 = top_predictions_2.mean()

print('Media de resrvas predichas en los mejores 200 pozos:\n')
print(f'\t- geo_data_0: {avg_top_pred_0:.2f} unidades')
# Comparar el umbral minimo de (111.1 unidades por pozo) 
if avg_top_pred_0 > ingreso_necesario_pozo/ingreso_barril:
    print("\tLa produccion media en los 200 pozos supera el minimo requerido de 111.1 unidades\n")
else: 
    print('\tLa produccion media en los top 200 pozos es inferior a 111.1 unidades\n') 

print(f'\t- geo_data_1: {avg_top_pred_1:.2f} unidades')
if avg_top_pred_1 > ingreso_necesario_pozo/ingreso_barril:
    print("\tLa produccion media en los 200 pozos supera el minimo requerido de 111.1 unidades\n")
else: 
    print('\tLa produccion media en los top 200 pozos es inferior a 111.1 unidades\n') 

print(f'\t- geo_data_2: {avg_top_pred_2:.2f} unidades')
if avg_top_pred_2 > ingreso_necesario_pozo/ingreso_barril:
    print("\tLa produccion media en los 200 pozos supera el minimo requerido de 111.1 unidades\n")
else: 
    print('\tLa produccion media en los top 200 pozos es inferior a 111.1 unidades\n') 
```

```python
# Calcular el beneficio potencial sin boostraping 

print(f'Beneficio potencial (sin bootstraping):\n')

#Geo_0 
total_product_top_0 = top_predictions_0.sum()
profit_0 = calc_profit(total_product_top_0)

print(f'\tgeo_data_0: ${profit_0:,.2f}')

#Geo_1
total_product_top_1 = top_predictions_1.sum()
profit_1 = calc_profit(total_product_top_1)

print(f'\tgeo_data_1: ${profit_1:,.2f}')

#Geo_2
total_product_top_2 = top_predictions_2.sum()
profit_2 = calc_profit(total_product_top_2)

print(f'\tgeo_data_2: ${profit_2:,.2f}')
```

En base a las cifras obenidas anteriormente, haria la selecion de geo_data_0 ya que es la que entrega un mayor beneficio. Sin embargo primero deberiamos hacer el analisis del riesgo asociado para poder tomar una decision mas acertada.

# 4. Calculo de riesgos y ganancias para cada region

```python
# funcion para aplicar la tecnica de bootstrapping para estrimar la distribucion del beneficio a partir de 200 pozos.

state = np.random.RandomState(45)

def bootstrap_profit(top_prediction, n_iterations = 1000):
    n_pozos = len(top_prediction)
    bootstrap_profits = []
    top_prediction = pd.Series(top_prediction)

    for i in range(n_iterations): 
        subsmaple = top_prediction.sample(frac = 1, replace=True, random_state=state) 
        total_predicted = subsmaple.sum()
        profit = calc_profit(total_predicted)
        bootstrap_profits.append(profit)

    bootstrap_profits = np.array(bootstrap_profits)
    mean_profit = np.mean(bootstrap_profits)
    ci = np.percentile(bootstrap_profits, [2.5,97.5]) 

    riesgo = (np.sum(bootstrap_profits < 0) / n_iterations ) * 100

    return mean_profit, ci, riesgo, bootstrap_profits
```

```python
print(f'Beneficio promedio (bootstrapping), Intervalo de confianza y reiesgo de perdida: \n')

# Calcular el riesgo y ganancia para geo_0
mean_profit_0, ci_0, riesgo_0, bootstrap_profits_0 = bootstrap_profit(top_predictions_0)

print('\tgeo_data_0\n')
print(f'\t\t- Beneficio promedio: ${mean_profit_0:,.2f}')
print(f'\t\t- Intervalo de confianza (95%): (${ci_0[0]:,.2f}, ${ci_0[1]:,.2f})')
print(f'\t\t- Riesgo de perdida: {riesgo_0:.2f}%\n')

# Calcular el riesgo y ganancia para geo_1
mean_profit_1, ci_1, riesgo_1, bootstrap_profits_1 = bootstrap_profit(top_predictions_1)

print('\tgeo_data_1\n')
print(f'\t\t- Beneficio promedio: ${mean_profit_1:,.2f}')
print(f'\t\t- Intervalo de confianza (95%): (${ci_1[0]:,.2f}, ${ci_1[1]:,.2f})')
print(f'\t\t- Riesgo de perdida: {riesgo_1:.2f}%\n')

# Calcular el riesgo y ganancia para geo_2
mean_profit_2, ci_2, riesgo_2, bootstrap_profits_2 = bootstrap_profit(top_predictions_0)

print('\tgeo_data_2\n')
print(f'\t\t- Beneficio promedio: ${mean_profit_2:,.2f}')
print(f'\t\t- Intervalo de confianza (95%): (${ci_2[0]:,.2f}, ${ci_2[1]:,.2f})')
print(f'\t\t- Riesgo de perdida: {riesgo_2:.2f}%\n')
```
